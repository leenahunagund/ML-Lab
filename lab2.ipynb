```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D as ax
```

```python
#generate smaple data
n_points=100
n_dimensions=3
data=np.random.rand(n_points,n_dimensions)
```

```python
#scatter points for 2d visualization
plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1],c='b',marker='o',label='dimension 1 vs dimension 2')
plt.xlabel('dimension 1')
plt.ylabel('dimension 2')
plt.title('scatter plot of 2d data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#box plot for visualizing distribution od each dimension
plt.figure(figsize=(8,6))
sns.boxplot(data=data)
plt.title('box plot of n-dimensional data')
plt.xlabel('dimension')
plt.ylabel('values')
plt.grid(True)
plt.show()
```

```python
#heat map for visualizing correlation btw dimensions
correlation_matrix=np.corrcoef(data,rowvar=False)
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f",
xticklabels=range(n_dimensions),yticklabels=range(n_dimensions))
plt.title('heatmap of correlation between dimensions')
plt.xlabel('dimensions')
plt.ylabel('dimensions')
plt.show()
```

```python
#contour plot for visualizing  3d data
if n_dimensions>=3:
  fig=plt.figure(figsize=(8,6))
  ax=fig.add_subplot(111,projection='3d')
  X,Y=np.meshgrid(data[:,0],data[:,1])
  Z=np.sin(np.sqrt(X**2+Y**2))
  ax.contour(X,Y,Z,50)
  ax.set_xlabel('dimension 1')
  ax.set_ylabel('dimension 2')
  ax.set_zlabel('dimension 3')
  plt.title("contour plot of 3d data")
  plt.show()

'''X = data[:, 0]
Y = data[:, 1]
Z = data[:, 2]

# Define grid
xi, yi = np.meshgrid(X, Y)


fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
zi=np.sin(np.sqrt(xi**2+yi**2))
ax.contour(xi, yi, zi, 50)  # Plot the contour
ax.set_xlabel('dimension 1')
ax.set_ylabel('dimension 2')
ax.set_zlabel('dimension 3')
plt.title("Contour Plot of 3D Data")
plt.show()'''
```

```python
#3D surface plot
if n_dimensions>=3:
    fig=plt.figure(figsize=(8,6))
    ax=fig.add_subplot(111,projection="3d")
    ax.plot_trisurf(data[:,0],data[:,1],data[:,2],cmap="viridis",edgecolor="none")
    ax.set_xlabel('dimension 1')
    ax.set_ylabel('dimension 2')
    ax.set_zlabel('dimension 3')
    plt.title("3d surface plot of 3d data")
    plt.show()
```
